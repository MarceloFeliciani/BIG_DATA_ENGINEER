### 3. Criar o DataFrame names_us para ler os arquivos no HDFS “/user/<nome>/data/names” com o seguinte schema:

#### nome: String
#### sexo: String
#### quantidade: Inteiro

```python
# importando a biblioteca

from pyspark.sql.types import *
```

```python
# Criando a estrutura

estrutura_lista = [
    StructField("nome", StringType()),
    StructField("sexo", StringType()),
    StructField("quantidade", IntegerType())
]

schema_names = StructType(estrutura_lista)


'''
PODERIA TER FEITO DESTA FORMA

schema_names = StructType([
    StructField("nome", StringType()),
    StructField("sexo", StringType()),
    StructField("quantidade", IntegerType())
])

'''
```

```python
names_us = spark.read.csv("/user/feliciani/data/names/", schema=schema_names)
```

### 4. Visualizar o Schema e os 5 primeiros registos do names_us

```python
print(names_us.printSchema())

names_us.show(5)

# Schema criado com cabeçalho e os dados tipados
```

### 5. Salvar o DataFrame names_us no formato orc no hdfs “/user/<nome>/names_us_orc”

```python
# HDFS já o padrão

names_us.write.orc("/user/feliciani/names_us_orc")
```

```python
!hdfs dfs -ls /user/feliciani/names_us_orc

# Criou o arquivo com 5 partições .
# Formato ORC salva por padrão com compressão SNAPPY
```
